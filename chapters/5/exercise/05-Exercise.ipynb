```python
pd.read_csv('https://api.statbank.dk/v1/data/FOLK1A/CSV?delimiter=Semicolon&OMR%C3%85DE=*&K%C3%98N=1%2C2&ALDER=*&CIVILSTAND=U%2CG&Tid=2020K4',sep=';')
```

```python
# 1. What is the change in pct of divorced danes from 2008 to 2020?
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('FOLK1A.csv', sep=';')
divorced = data.loc[(data['OMRÅDE'] == 'Hele landet') &  (data['CIVILSTAND']=='Fraskilt')].iloc[:]
display(divorced)

# bar plot
plt.plot(divorced['TID'], divorced['INDHOLD'], color='r', label='Divorced')
plt.xticks(rotation=45, horizontalalignment='right', fontweight='light')
plt.ylabel("People")
plt.xlabel("Year")
plt.title("Divorced")
plt.legend()
plt.show()

divorced_2008 = divorced.loc[data['TID'] == '2008K4'].iloc[0][4]
divorced_2020 = divorced.loc[data['TID'] == '2020K4'].iloc[0][4]
change_in_pct = (divorced_2020-divorced_2008)/divorced_2008*100
print("Divorced in 2008: {}\nDivorced in 2020: {}\nChange: {}%\n".format(divorced_2008, divorced_2020, change_in_pct))
```

```python
 # 2. Which of the 5 biggest cities has the highest percentage of 'Never Married' in 2020?
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('FOLK1A.csv', sep=';')

never_married = data.loc[(data['OMRÅDE'] != 'Hele landet') &  (data['CIVILSTAND']=='Ugift') & (data['TID']=='2020K4')].iloc[:]

display(never_married)

# bar plot
plt.bar(never_married['OMRÅDE'], never_married['INDHOLD'], color='r', label='Never married', align='center')
plt.xticks(rotation=45, horizontalalignment='right', fontweight='light')
plt.ylabel("People")
plt.xlabel("Area")
plt.title("Never married")
plt.legend()
plt.show()
```

```python
# 3. Show a bar chart of changes in marrital status in Copenhagen from 2008 till now
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('FOLK1A.csv', sep=';')

married = data.loc[(data['OMRÅDE']=='Hele landet') & (data['CIVILSTAND']=='Gift/separeret')].iloc[:,2:]
never_married = data.loc[(data['OMRÅDE']=='Hele landet') & (data['CIVILSTAND']=='Ugift')].iloc[:,2:]
divorced = data.loc[(data['OMRÅDE']=='Hele landet') & (data['CIVILSTAND']=='Fraskilt')].iloc[:,2:]

display(married)
display(never_married)
display(divorced)

# bar plot with three bars
plt.bar(never_married['TID'], never_married['INDHOLD'], color='r', label='Never married', align='center')
plt.bar(married['TID'], married['INDHOLD'], color='g', label='Married', align='center')
plt.bar(divorced['TID'], divorced['INDHOLD'], color='b', label='Divorced')
plt.xticks(rotation=45, horizontalalignment='right', fontweight='light')
plt.ylabel("People")
plt.xlabel("Year")
plt.title("People marrital status in DK")
plt.legend()
plt.show()
```

```python
# 4. Show 2 plots in same figure: 'Married' and 'Never Married' for all ages in DK in 2020 (Hint: x axis is age from 0-125, y axis is how many people in the 2 categories). Add lengend to show names on graphs
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('FOLK1A.csv', sep=';')
# iloc can be used to slice column or row in a dataframe (integer-location based indexing)
# loc to make condition for the data frame row or columns with labels (label based data selecting method)
married = data.loc[(data['OMRÅDE']=='Hele landet') & (data['CIVILSTAND']=='Gift/separeret')].iloc[:,2:] # (gets all rows and gets the column TID and INDHOLD)
never_married = data.loc[(data['OMRÅDE']=='Hele landet') & (data['CIVILSTAND']=='Ugift')].iloc[:,2:]

display(married)
display(never_married)


# line plot with two graphs
plt.plot(married['TID'], married['INDHOLD'], color='g', label='Married')
plt.plot(never_married['TID'], never_married['INDHOLD'], color='r', label='Never married')
plt.xticks(rotation=45, horizontalalignment='right', fontweight='light')
plt.ylabel("People")
plt.xlabel("Year")
plt.title("Married vs Never married in DK")
plt.legend()
plt.show()
```
