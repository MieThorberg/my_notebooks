```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./titanic_train.csv')

# set index of the dataframe to passenger id
df.set_index('PassengerId')

# sort the dataframe by 'fare' (price of ticket)
df.sort_values('Fare', ascending=True).head()

# write a function to reformat a name to only contain "firstname lastname" (look for current format in the name column)
# def reformat_name(name):
#     _name = name.split(',')
#     if _name[1].__contains__("Mrs."):
#         _name[1] = _name[1].replace("Mrs.", "")
#     if _name[1].__contains__("Miss."):
#         _name[1] = _name[1].replace("Miss.", "")
#     if _name[1].__contains__("Mr."):
#         _name[1] = _name[1].replace("Mr.", "")
#
#     return _name[1] + " " + _name[0]

def reformat_name(name):
    first_name = name.split(', ')[1].split(' ')[1].strip()
    last_name = name.split(',')[0].strip()

    return f'{first_name} {last_name}'

df['Name'] = df['Name'].apply(reformat_name)



# Get only last name by the use of str method split() on comma.
def get_only_lastname(name):
    _name = name.split(' ')
    return _name[1]
df['Name'] = df['Name'].apply(get_only_lastname)
df.head()


# Get only those passengers that had one of these ticket numbers: `[350406,248706,382652,244373,345763,2649,239865]`
tickets = ['350406','248706','382652','244373','345763','2649','239865']
df_tickets = df[df["Ticket"].isin(tickets)]

# Remove all rows without and age
df_tickets = df_tickets[df_tickets["Age"].notna()]
df_tickets

```
